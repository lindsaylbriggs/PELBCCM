```python
import os
import numpy as np
import pandas as pd
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image_directory = 'drive/MyDrive/CCM/Test_Datasets/'

```

```python
data = []
for file in os.listdir(image_directory):
  if file.endswith('.png'):
    imgpath = os.path.join(image_directory, file)
    img = Image.open(imgpath)
    img = img.resize((128, 128))
    img_array = np.array(img)
    data.append({'name': imgpath, 'Image': img_array, 'label': None})
  else:
    pass
```

```python
df = pd.DataFrame(data)
df
```
